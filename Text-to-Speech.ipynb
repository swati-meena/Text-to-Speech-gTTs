### GTTS - Google Text to Speech

### Text to Speech conversion using GTTS

```python
# pip install gtts
```

```python
from gtts import gTTS
from IPython.display import Audio

text_to_speech = gTTS('''gTTS (Google Text-to-Speech) is a Python library that allows users to convert text into speech 
                        using Google's Text-to-Speech API. It is widely used for creating speech output from text in a variety of languages,
                        making it useful for applications such as chatbots, virtual assistants, or accessibility tools.''',lang='hi',tld='com')
text_to_speech.save('text_to_speech_gtts.wav')
sound_file = 'text_to_speech_gtts.wav'
Audio(sound_file, autoplay= False)
```

### Text to Speech conversion using pyttsx3

#### Changing the gender using pyttsx3

```python
# pip install pyttsx3
```

```python
import pyttsx3
from IPython.display import Audio

text = '''gTTS (Google Text-to-Speech) is a Python library that allows users to convert text into speech using Google's Text-to-Speech API. 
          It is widely used for creating speech output from text in a variety of languages,
          making it useful for applications such as chatbots, virtual assistants, or accessibility tools.'''

audio = pyttsx3.init()
audio.setProperty('rate', 150)
audio.setProperty('volume', 0.8)

# change the voices
voice = audio.getProperty('voices')

# 0 for male and 1 for female
audio.setProperty('voice', voice[0].id)      # for male voice   
#audio.setProperty('voice', voice[1].id)       # for female voice

# text-to speech conversion
audio.say(text)

# save the audio file
audio.save_to_file(text, 'test_male_Voice.mp3')
#audio.save_to_file(text, 'test_female_Voice.mp3')

audio.runAndWait()
```
